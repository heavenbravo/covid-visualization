```python
df1 = df.shift(periods=1, axis=1, fill_value = 0)
df1
```

```python
Dayy = df[days] - df1[days]
```

```python
Dayy.loc['total'] = Dayy.sum(axis = 0)
Dayy
```

```python
Dayy.loc['total'].plot(kind = 'area', figsize = (20, 10))
plt.title('Trending of Daily Increase of the World')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
## Look at US indivisually
df_us = Dayy.loc['US', days]
df_us.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Daily US Cases ')
plt.ylabel('Number of cases')

plt.show()
```
